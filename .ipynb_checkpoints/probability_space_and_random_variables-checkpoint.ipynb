## Вероятностное пространство и случайные величины

```python
import math as m
import numpy as np
from tqdm.notebook import tqdm
import random
import matplotlib.pyplot as plt
import scipy.stats as sps
```

### 1 задача

$\Omega$ является площадью квадрата, событие $A = \{$точка попала в площадь круга, находящегося в квадрате$\}$.

Вероятность того, что случайно брошенная точка окажется в круге, равна отношению площади круга к площади квадрата. Радиус круга равен единице.

$P(A)=\dfrac{\lambda(A)}{\lambda(\Omega)}=\dfrac{1\cdot \pi}{4\cdot 4}=\dfrac{\pi}{16} \approx 0.196$

```python
side_square = 4
radius_circle = 1
n = 1000000
success = 0

for _ in tqdm(range(n)):
    x = np.random.uniform(0, side_square)
    y = np.random.uniform(0, side_square)
    if (x - radius_circle)**2 + (y - radius_circle)**2 <= radius_circle**2:
        success += 1

P = success / n
P
```

### 2 задача

Пусть событие $А = \{$ вынули синий шар $\}, B_1 = \{$выбрана первая урна$\}$ и $B_2 = \{$выбрана вторая урна$\}$.

Тогда $P(A) = P(A|B_1) + P(A|B_2)$.

Вероятность выбора любой урны одинакова, то есть $\dfrac{1}{2}$.

$P(A|B_1) = \dfrac{P(AB_1)}{B_1} = \dfrac{\dfrac{3}{5} \cdot \dfrac{1}{2}}{\dfrac{1}{2}} = \dfrac{3}{5}$

$P(A|B_2) = \dfrac{P(AB_2)}{B_2} = \dfrac{1}{3}$

Вероятность того, что вынутый шар будет синим, равна $\dfrac{3}{5} + \dfrac{1}{3} = \dfrac{7}{15} \approx 0.467$

```python
n = 100000
urn1 = np.array([0] * 2 + [1] * 3)
urn2 = np.array([0] * 4 + [1] * 2)
success = 0

for _ in tqdm(range(n)):
    chosen_urn = urn1 if np.random.random() < 0.5 else urn2
    chosen_ball = np.random.choice(chosen_urn)
    if chosen_ball == 1:
        success += 1

# Вычисление вероятности
probability = success / n
probability
```

### 3 задача

    
 $\Omega$ состоит из всевозможных исходов вида $(a_1,a_2,a_3)$, где $a_i\in\{$белый шар, чёрный шар$\}$. Достать шары двух цветов возможны в двух вариантах: 
 - вытянуть 2 белых и 1 чёрный шар
 - вытянуть 1 белый и 2 чёрных шара

$P($оба цвета$)=P($2 белых$,$1 черный$)+P($1 белый$,$2 черных$)$


$P($2 белых$,$1 черный$) = \dfrac{C^{2}_{3}\cdot C^{1}_5}{C^{3}_{8}} = \dfrac{15}{56}$

$P($1 белый$,$2 черных$) = \dfrac{C^{1}_{3}\cdot C^{2}_5}{C^{3}_{8}} = \dfrac{15}{28}$

И тогда вероятность обоих случаев равна $\dfrac{15}{56} + \dfrac{15}{28} = \dfrac{45}{56} \approx 0.804$

```python
def choose(n,k):
    return m.factorial(n) // m.factorial(n-k) // m.factorial(k)

a = 3
b = 5

choose(a,2) * choose(b, 1) / choose(a+b, 3) + choose(a,1) * choose(b, 2) / choose(a+b, 3)
```

```python
n = 100000
urn = np.array([0] * a + [1] * b)

success = 0

for _ in tqdm(range(n)):
    choice = np.random.choice(urn, 3, replace=False)
    if sum(choice) in [1, 2]:
        success += 1

success / n
```

### 4 задача

Так как n велико, воспользуемся интегральной формулой Муавра-Лапласа:

$P(a \leq X \leq b) \approx \Phi\left(\frac{b - np}{\sqrt{npq}}\right) - \Phi\left(\frac{a - np}{\sqrt{npq}}\right)=$

$n = 100, a = 70, b = 80, p = 0.75, q = 1 - p = 0.25$

$x_1 =  \dfrac{80-100 \cdot 0.75}{\sqrt{100 \cdot 0.75 \cdot 0.25}} \approx 1.154$

$x_2 =  \dfrac{70-100 \cdot 0.75}{\sqrt{100 \cdot 0.75 \cdot 0.25}} \approx -1.154$

$ = \Phi(1.154) - \Phi(-1.154) = 2\Phi(1.154) \approx 0.7498$ (по таблице значений)

```python
n = 50000000
shots = 100
P_hit = 0.75
hits_range = range(70, 81)

success = 0
for _ in tqdm(range(n)):
    hits = np.random.binomial(shots, P_hit)
    if hits in hits_range:
        success += 1

# Вычисление вероятности
P = success / n
P
```

1. Необходимо найти параметр $c$ так, чтобы функция была непрерывной.

Функция распределения должна быть непрерывной в каждой точке, включая $x = -1$. Значит, левый и правый пределы функции в точке$x = -1$ должны совпадать. По определению функции, для $ x \leq -1 $, $F_{\varepsilon}(x) = 0$.

Рассмотрим значение функции сразу после $x = -1$, то есть в интервале $ -1 < x \leq \frac{1}{3} $, которое задаётся как $\frac{3}{4}x + c$. При $x = -1 $ это выражение должно также равняться 0, чтобы функция была непрерывной:

$\frac{3}{4}(-1) + c = 0 $

$-\frac{3}{4} + c = 0$

$c = \frac{3}{4}$

2.
$$
f_{\varepsilon}(x) = \begin{cases} 
0, & x \leq -1 \\
\frac{3}{4}, & -1 < x \leq \frac{1}{3} \\
0, & x > \frac{1}{3} 
\end{cases}$$

3.
Математическое ожидание $ M(X)$ для непрерывной случайной величины определяется как $ M(X) = \int_{-\infty}^{\infty} x f(x) dx$. Подставим в формулу: $ M(X) = \int_{-1}^{\frac{1}{3}} x \cdot \frac{3}{4} dx = -\frac{1}{3} $

Дисперсия $ D(X) = M(X^2) - (M(X))^2 $, где $ M(X^2) = \int_{-\infty}^{\infty} x^2 f(x) dx $

$ M(X^2) = \int_{-1}^{\frac{1}{3}} \frac{3}{4} x^2 dx = \frac{3}{4} \left[ \frac{x^3}{3} \right]_{-1}^{\frac{1}{3}} = \frac{21}{81} \approx 0.259$

$ D(X) = \frac{21}{81} - \left(-\frac{1}{3}\right)^2 \approx 0.148 $

Медиана $ F(Me) = 0.5 $

$F(X)$ для $-1 < x \leq \frac{1}{3} $ равна $ \frac{3}{4}x + \frac{3}{4} $. Поэтому медиана $ Me $ решает уравнение $ \frac{3}{4}Me + \frac{3}{4} = 0.5 $

$ Me = -\frac{1}{3} $

4. $ P(0 < \xi < \frac{1}{3}) = \int_{0}^{\frac{1}{3}} f(x) dx = \int_{0}^{\frac{1}{3}} \frac{3}{4}dx= \frac{3}{4} \left[ x \right]_{0}^{\frac{1}{3}} = 0.25$

```python
x = np.linspace(-2, 1, 400)

F_x = np.piecewise(x, [x <= -1, (x > -1) & (x <= 1/3), x > 1/3], [0, lambda x: 3/4*x + 0.75, 1])
f_x = np.piecewise(x, [x <= -1, (x > -1) & (x <= 1/3), x > 1/3], [0, 0.75, 0])

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, F_x, label="F(x)", color="blue")
plt.title("F(x)")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, f_x, label="f(x)", color="red")
plt.title("f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

### 2 задача

Распределение числа испытаний до первого отказа - это геометрическое распределение: $ P(X = k) = (1 - p)^{k-1} \cdot p $

Математическое ожидание: $M(X) = \frac{1}{p}$

Дисперсия: $D(X) = \frac{1 - p}{p^2}$.

Для $p = 0.1$:

- $M(X) = \frac{1}{0.1} = 10$
- $D(X) = \frac{1 - 0.1}{0.1^2} = 90$

Кумулятивная вероятность того, что произойдет k или менее неудач до первого успеха, находится по формуле $ P(X \leq k) = 1 - (1 - p)^k $.

Подберём минимальное значение `k`, при котором $ P(X \leq k) > 0.8 $.

$ P(X \leq 1) = 1 - (1 - 0.1)^1 = 1 - 0.9 = 0.1 $

$ P(X \leq 10) = 1 - (1 - 0.1)^{10} = 1 - 0.9^{10} \approx 0.65132$

$ P(X \leq 16) = 1 - (1 - 0.1)^{16} = 1 - 0.9^{16} \approx 0.81469$

Таким образом, нужно провести минимум 16 испытаний, чтобы вероятность отказа превысила 0.8.

```python
p_failure = 0.1

geom_dist = sps.geom(p_failure)

print(f"Мат ожидание: {geom_dist.mean()}")
print(f"Дисперсия: {geom_dist.var()}")
```

```python
k = 1
while geom_dist.cdf(k) <= 0.8:
    k += 1
    print("k = " + str(k) + ": " + str(round(geom_dist.cdf(k), 4)))

print(f"Минимальное количество испытаний, чтобы вероятность отказа превысила 0.8: {k}")
```

```python
n_series = 500000000
p_failure = 0.1

first_failures = np.random.geometric(p_failure, n_series)

for k in range(1, 100):
    print("k = " + str(k) + ": " + str(round(np.mean(first_failures <= k), 4)))
    if np.mean(first_failures <= k) > 0.8:
        break

print(f"Минимальное значение k, при котором вероятность отказа превышает 0.8: {k}")
```

### 3 задача

Вероятность того, что значение величины отклонится от среднего меньшее чем на $\delta$, равна $P(|X - a| < \delta) = 2 \Phi\left(\frac{\delta}{\sigma} \right) $

где:
- $\Phi$ — функция стандартного нормального распределения,
- $\sigma$ — стандартное отклонение,
- $\delta$ — максимальное отклонение от среднего значения.

$ \sigma = \sqrt{\sigma^2} = \sqrt{0.1} \approx 0.3162 $

$ P(|X - 5| < 0.4) = 2 \Phi\left(\frac{0.4}{0.3162}\right) $

$\Phi$ от 1.26 по таблице значений равно 0.3962, поэтому $ P(|X - 5| < 0.4) = 2 \cdot 0.3962 \approx 0.7924 $.

То есть, вероятность того, что диаметр шарика отклонится от среднего на не более чем 0.4 мм, составляет примерно 0.7924.

```python
mu = 5
sigma = np.sqrt(0.1)
delta = 0.4
n = 50000000

successes = 0
for _ in tqdm(range(n)):
    diameter = np.random.normal(mu, sigma)
    if mu - delta <= diameter <= mu + delta:
        successes += 1

P = successes / n
P
```

### 4 задача

Решим с помощью интегральной формулы Муавра-Лапласа:

$P(a \leq X \leq b) \approx \Phi\left(\frac{b - np}{\sqrt{npq}}\right) - \Phi\left(\frac{a - np}{\sqrt{npq}}\right)=$

$n = 12000, a = 1900, b = 2150, p = \dfrac{1}{6}, q = 1 - p = \dfrac{5}{6}$

$x_1 =  \dfrac{-100}{100/\sqrt{6}} = -\sqrt{6}$

$x_2 =  \dfrac{150}{100/\sqrt{6}} = \dfrac{3 \cdot \sqrt{6}}{2}$

$ = \Phi(\dfrac{3 \cdot \sqrt{6}}{2}) - \Phi(-\sqrt{6}) = \Phi(3.67) + \Phi(2.45) \approx 0.99$ (по таблице значений)

```python
n = 50000
rolls = 12000

def simulate_dice_rolls(n):
    return sum(1 for _ in range(n) if random.randint(1, 6) == 1)

success = 0
for _ in tqdm(range(n)):
    count_ones = simulate_dice_rolls(rolls)
    if 1900 <= count_ones <= 2150:
        success += 1

# Вычисление вероятности
P = success / n
P
```
